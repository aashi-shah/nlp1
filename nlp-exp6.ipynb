```python
# Import necessary libraries
import nltk
import spacy
```

```python
# Download necessary NLTK packages if not already downloaded
nltk.download('punkt', quiet=True)
nltk.download('averaged_perceptron_tagger', quiet=True)
```

```python
# Load spaCy's small English model
nlp = spacy.load("en_core_web_sm")
```

```python
# Function to perform POS tagging using NLTK
def pos_tag_nltk(text):
    words = nltk.word_tokenize(text)
    pos_tags = nltk.pos_tag(words)
    return pos_tags

# Function to perform POS tagging using spaCy
def pos_tag_spacy(text):
    doc = nlp(text)
    pos_tags = [(token.text, token.pos_) for token in doc]
    return pos_tags
```

```python
# Function to read text from a file
def read_text_file(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            return file.read()
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
        return None
```

```python
# Function to display POS tags
def display_pos_tags(pos_tags, library_name):
    print(f"\nPOS tagging using {library_name}:\n")
    for word, tag in pos_tags:
        print(f"{word} -> {tag}")
```
